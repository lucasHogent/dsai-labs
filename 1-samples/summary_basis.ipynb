### Cheatsheet voor Basisbegrippen en Steekproeven

#### Basisbegrippen
1. **Variabele**: Een eigenschap of kenmerk dat kan variëren tussen verschillende eenheden of individuen binnen een studie. Voorbeeld: Leeftijd, geslacht, inkomen.

2. **Waarde**: Een specifieke eigenschap of kenmerk van een variabele. Voorbeeld: Als de variabele leeftijd is, kunnen de waarden 25, 30, 45, etc. zijn.

#### Meetniveau
1. **Nominaal**:
   - **Kenmerken**: Categorieën zonder inherente volgorde.
   - **Voorbeeld**: Geslacht (man, vrouw), kleur (rood, blauw, groen).

2. **Ordinaal**:
   - **Kenmerken**: Categorieën met een logische volgorde, maar zonder vaste afstanden tussen de categorieën.
   - **Voorbeeld**: Opleidingsniveau (laag, middelbaar, hoog), klanttevredenheid (slecht, matig, goed).

3. **Interval**:
   - **Kenmerken**: Gelijkaardige afstanden tussen waarden, geen absoluut nulpunt.
   - **Voorbeeld**: Temperatuur in Celsius of Fahrenheit, jaartallen.

4. **Ratio**:
   - **Kenmerken**: Gelijkaardige afstanden tussen waarden met een absoluut nulpunt.
   - **Voorbeeld**: Lengte, gewicht, inkomen.

#### Populatie en Steekproef
1. **Populatie**: De gehele groep individuen of eenheden die je wilt bestuderen.
   - **Voorbeeld**: Alle inwoners van Nederland.

2. **Steekproef**: Een deelverzameling van de populatie die wordt bestudeerd om conclusies te trekken over de gehele populatie.
   - **Voorbeeld**: 1000 willekeurig gekozen Nederlanders voor een enquête.

3. **Steekproefkader**: De lijst of database waaruit een steekproef wordt getrokken.
   - **Voorbeeld**: Een lijst met alle adressen in Nederland.

#### Aselecte vs. Selecte Steekproef
1. **Aselecte steekproef**: Elk lid van de populatie heeft een gelijke kans om geselecteerd te worden.
   - **Voorbeeld**: Willekeurige nummerselectie uit een telefoongids.

2. **Selecte steekproef**: Niet elk lid van de populatie heeft een gelijke kans om geselecteerd te worden, vaak gebaseerd op specifieke criteria.
   - **Voorbeeld**: Een enquête onder alleen universiteitsstudenten.

#### Representativiteit van een Steekproef
- Een steekproef is representatief als de kenmerken van de steekproef overeenkomen met de kenmerken van de populatie, zodat de resultaten generaliseerbaar zijn.

#### Types Steekproeffouten
1. **Steekproeffout**:
   - **Kenmerken**: Variaties die ontstaan door het gebruik van een steekproef in plaats van de volledige populatie.
   - **Voorbeeld**: Verschillende resultaten in meerdere steekproeven van dezelfde populatie.

2. **Selectie Bias**:
   - **Kenmerken**: Systematische fouten die ontstaan wanneer sommige leden van de populatie minder kans hebben om in de steekproef te komen.
   - **Voorbeeld**: Alleen telefonisch enquêteren kan mensen zonder telefoon uitsluiten.

3. **Respons Bias**:
   - **Kenmerken**: Fouten die ontstaan door verschillen in de antwoorden van respondenten.
   - **Voorbeeld**: Mensen geven sociaal wenselijke antwoorden in plaats van hun werkelijke mening.

#### Voorstellen tot Verbetering van de Steekproefmethode
- **Gebruik aselecte steekproefmethoden**: Zoals eenvoudige willekeurige steekproef, gestratificeerde steekproef, of clustersteekproef om representativiteit te verhogen.
- **Vergroot steekproefgrootte**: Om de precisie van de schattingen te verbeteren en de steekproeffout te verminderen.
- **Garandeer responsrepresentativiteit**: Door non-respons te verminderen en ervoor te zorgen dat de steekproef alle subgroepen binnen de populatie adequaat vertegenwoordigt.

#### Meetniveau voor een Gegeven Variabele
- Bepaal het meetniveau door de aard van de data te evalueren: Kijk naar de eigenschappen van de variabele en bepaal of er een rangorde, gelijke afstanden, en een absoluut nulpunt aanwezig zijn.

Door deze cheatsheet te gebruiken, kun je de basisbegrippen en meetniveaus in onderzoek begrijpen en toepassen, en verbeteringen aanbrengen in steekproefmethoden om betrouwbaarheid en representativiteit te vergroten.

# Steekproeffouten (Sampling Errors) en Niet-Steekproeffouten (Non-Sampling Errors)

#### Steekproeffouten (Sampling Errors)
**Definitie**: Fouten die ontstaan doordat een steekproef in plaats van de volledige populatie wordt gebruikt. Deze fouten kunnen niet volledig worden geëlimineerd, maar ze kunnen wel worden verminderd door bijvoorbeeld de steekproefgrootte te vergroten.

1. **Toevallige steekproeffout** Accidental:
   - **Kenmerken**: Fouten die ontstaan door willekeurige variatie tussen steekproeven.
   - **Voorbeeld**: Bij het trekken van verschillende steekproeven uit dezelfde populatie kunnen de resultaten licht variëren.

2. **Systematische steekproeffout** Systematic:
   - **Kenmerken**: Fouten die ontstaan door een fout in de steekproefmethode waardoor bepaalde groepen over- of ondervertegenwoordigd zijn.
   - **Voorbeeld**: Een steekproef die alleen uit stedelijke gebieden komt, terwijl de populatie zowel stedelijke als landelijke gebieden omvat.

**Voorbeeld**: Als je een enquête uitvoert onder 1000 willekeurig geselecteerde mensen in Nederland, kunnen de resultaten afwijken van de werkelijke waarden in de gehele bevolking door puur toeval.

#### Niet-Steekproeffouten (Non-Sampling Errors)
**Definitie**: Fouten die ontstaan door andere factoren dan het trekken van een steekproef. Deze fouten kunnen optreden in alle fasen van het onderzoek en zijn vaak systematisch van aard.

1. **Selectie Bias**:
   - **Kenmerken**: Fouten die ontstaan wanneer de methode van steekproeftrekking bepaalde delen van de populatie uitsluit.
   - **Voorbeeld**: Als alleen mensen met een telefoon worden gebeld, worden mensen zonder telefoon uitgesloten.

2. **Niet-Respons Bias**:
   - **Kenmerken**: Fouten die ontstaan wanneer bepaalde groepen in de steekproef niet antwoorden of niet kunnen worden bereikt.
   - **Voorbeeld**: Een groot aantal mensen reageert niet op een schriftelijke enquête, en degenen die wel reageren verschillen systematisch van degenen die niet reageren.

3. **Meetfouten**:
   - **Kenmerken**: Fouten die ontstaan door onnauwkeurigheden in de manier waarop gegevens worden verzameld.
   - **Voorbeeld**: Onjuiste antwoorden door misverstanden over vragen of fouten bij het invoeren van gegevens.

4. **Verwerkingsfouten**:
   - **Kenmerken**: Fouten die ontstaan bij het verwerken en analyseren van gegevens.
   - **Voorbeeld**: Fouten bij het coderen van open antwoorden of bij het invoeren van data in een database.

5. **Interviewerbias**:
   - **Kenmerken**: Fouten die ontstaan door de invloed van de interviewer op de respondenten.
   - **Voorbeeld**: Een interviewer stelt leidende vragen die de antwoorden van respondenten beïnvloeden.

**Voorbeeld**: Als bij een enquête over internetgebruik alleen mensen worden bevraagd die regelmatig internet gebruiken, ontstaat een niet-respons bias omdat mensen die zelden of nooit internet gebruiken niet worden vertegenwoordigd.

#### Oplossingen om Steekproeffouten en Niet-Steekproeffouten te Verminderen

1. **Steekproeffouten**:
   - **Vergroot de steekproefgrootte**: Een grotere steekproef vermindert de toevallige variatie.
   - **Gebruik gestratificeerde steekproefmethoden**: Verdeelt de populatie in subgroepen en neemt uit elk subgroep een aselecte steekproef om systematische fouten te verminderen.

2. **Niet-Steekproeffouten**:
   - **Garandeer representativiteit**: Gebruik methoden zoals gestratificeerde of clustersteekproeven.
   - **Verminder non-respons**: Stuur herinneringen en zorg voor meerdere manieren om te reageren (bijvoorbeeld telefonisch en online).
   - **Train interviewers**: Zorg voor consistente en neutrale vraagstelling.
   - **Verfijn meetinstrumenten**: Gebruik duidelijke en begrijpelijke vragen en test deze vooraf.
   - **Controleer en verifieer data**: Pas kwaliteitscontroles toe tijdens het data-invoerproces.

Door deze maatregelen kunnen zowel steekproeffouten als niet-steekproeffouten worden geminimaliseerd, wat leidt tot betrouwbaardere en nauwkeurigere onderzoeksresultaten.

# Wat is Bias?

**Bias** is een systematische fout of vertekening in een onderzoeksproces die leidt tot onnauwkeurige of vertekende resultaten. Het kan op verschillende manieren ontstaan en kan zowel bewust als onbewust plaatsvinden. Bias beïnvloedt de validiteit van de resultaten en kan ertoe leiden dat conclusies niet representatief zijn voor de werkelijke situatie.

#### Types van Bias

1. **Selectie Bias**:
   - **Kenmerken**: Treedt op wanneer de methode van steekproeftrekking ervoor zorgt dat bepaalde delen van de populatie niet evenredig vertegenwoordigd zijn.
   - **Voorbeeld**: Als een onderzoek naar de gezondheid van ouderen alleen deelnemers rekruteert via internet, worden ouderen zonder internettoegang uitgesloten.

2. **Niet-Respons Bias**:
   - **Kenmerken**: Treedt op wanneer een significante groep binnen de steekproef niet reageert, en deze groep systematisch verschilt van degenen die wel reageren.
   - **Voorbeeld**: Bij een schriftelijke enquête reageren alleen mensen die erg geïnteresseerd zijn in het onderwerp, terwijl de rest van de steekproef niet reageert.

3. **Meetfouten (Measurement Bias)**:
   - **Kenmerken**: Treedt op wanneer de methoden of instrumenten die worden gebruikt om gegevens te verzamelen systematisch onnauwkeurig zijn.
   - **Voorbeeld**: Een vragenlijst over dieetgewoonten die complexe en moeilijke vragen stelt, kan leiden tot onnauwkeurige antwoorden.

4. **Interviewer Bias**:
   - **Kenmerken**: Treedt op wanneer de interviewer onbewust of bewust de antwoorden van de respondenten beïnvloedt door de manier van vragen stellen of reageren op antwoorden.
   - **Voorbeeld**: Een interviewer die glimlacht bij bepaalde antwoorden kan respondenten aanmoedigen om meer van zulke antwoorden te geven.

5. **Verwerkingsfouten (Processing Bias)**:
   - **Kenmerken**: Treedt op tijdens de verwerking van gegevens, zoals bij het coderen, invoeren of analyseren van data.
   - **Voorbeeld**: Fouten bij het invoeren van data kunnen leiden tot verkeerde conclusies.

6. **Publicatiebias (Publication Bias)**:
   - **Kenmerken**: Treedt op wanneer de kans op publicatie van onderzoeksresultaten afhankelijk is van de aard en richting van de resultaten.
   - **Voorbeeld**: Onderzoeken met significante of positieve resultaten worden vaker gepubliceerd dan onderzoeken met niet-significante of negatieve resultaten.

#### Voorbeelden van Bias en Hun Effecten

- **Voorbeeld**: Een medisch onderzoek naar de effectiviteit van een nieuw medicijn rekruteert alleen patiënten die gezond genoeg zijn om naar een ziekenhuis te komen voor testen. Dit kan selectie bias introduceren, omdat ernstig zieke patiënten worden uitgesloten, wat kan leiden tot een overschatting van de effectiviteit van het medicijn.
- **Effecten**: Bias kan leiden tot verkeerde conclusies, zoals het onderschatten of overschatten van het effect van een behandeling, het verkeerd inschatten van de prevalentie van een probleem, of het trekken van onjuiste conclusies over causale relaties.

#### Hoe Bias te Verminderen

1. **Gebruik aselecte steekproeftrekking**: Zorg ervoor dat elke eenheid in de populatie een gelijke kans heeft om in de steekproef te komen.
2. **Vergroot de respons**: Gebruik herinneringen en meerdere manieren om te reageren om de non-respons te verminderen.
3. **Train interviewers**: Zorg voor consistente en neutrale vraagstelling om interviewer bias te verminderen.
4. **Verfijn meetinstrumenten**: Test en verbeter vragenlijsten om meetfouten te verminderen.
5. **Blinde onderzoeken**: Gebruik blinde of dubbelblinde methoden waarbij de onderzoekers en/of deelnemers niet weten welke interventie wordt gegeven, om interviewer en beoordelaar bias te verminderen.
6. **Publicatie van alle resultaten**: Publiceer alle onderzoeksresultaten, ongeacht de aard en richting van de uitkomsten, om publicatiebias te verminderen.

Bias is een cruciale factor om rekening mee te houden bij het ontwerpen, uitvoeren en interpreteren van onderzoek om de validiteit en betrouwbaarheid van de resultaten te waarborgen.
