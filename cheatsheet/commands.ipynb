Load lib

```python
# Importing the necessary packages
import numpy as np                                  # "Scientific computing"
import scipy.stats as stats                         # Statistical tests

import pandas as pd                                 # Data Frame
from pandas.api.types import CategoricalDtype

import matplotlib.pyplot as plt                     # Basic visualisation
from statsmodels.graphics.mosaicplot import mosaic  # Mosaic diagram
import seaborn as sns                               # Advanced data visualisation
from scipy.special import binom # binomial-function
```

Read csv

```python
list = pd.read_csv('../data/titanic.csv', delimiter=',')
list.head()

# Parse date
# list = pd.read_csv('../data/gasoline.csv', delimiter=';',index_col=0, parse_dates=True)
# list.head()
 
```

Change column data type

```python
# Describe the variable Survived -> is considered to be quantitative
print(list.Survived.describe())
# Convert to a categorical variable
list.Survived = list.Survived.astype('category')
# Ask to describe once more -> now it is considered to be qualitative
print(list.Survived.describe())

# Change to ordinal

print(list.Embarked.unique())

embarked_type = CategoricalDtype(categories=['S', 'C', 'Q'], ordered=True)
list.Embarked = list.Embarked.astype(embarked_type)
list.Embarked.describe()


```

Add column

```python
list["Survived_boolean"] = list["Survived"] == 1
list.head()

# Ratio
list["ratio_age_fare"] = list["Age"] / list["Fare"]

# Map values to other 
mapping_dict = {'male' : 0, 'female' : 1}
list['Sex'] = list['Sex'].map(mapping_dict)

# Other way to map
def age_to_category(age):
    if age < 12:
        return "child"
    if age < 18:
        return "teen"
    return "adult"

list['AgeCategory'] = list['Age'].map(age_to_category)

list.head()
```

Clean table

```python
# drop column
list = list.drop("PassengerId", axis="columns")

# Drop any row that has at least one missing value
list = list.dropna()

# Drop only the rows that consist of nothing but missing values
list = list.drop(how="all")

# Fill missing values
avg_age = list['Age'].mean()
print(f"(Rounded) Average age of passengers: {round(avg_age)}")
titanic = list.fillna(value={'Age' : avg_age})
print("We should now confirm that the 'Age' column no longer has missing values")
titanic.info()
```
